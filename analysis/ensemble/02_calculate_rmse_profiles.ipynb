```python
### Prepare to open data
# rank_dirs = [Path('/projects/wakedynamics/orybchuk/ldm-3d/logs/2022-12-26T07-28-14_rank_geo_raaw_kl1_0000/images/test/postprocessed/'),
#                 Path('/projects/wakedynamics/orybchuk/ldm-3d/logs/2022-12-26T07-28-14_rank_geo_raaw_kl1_0010/images/test/postprocessed/'),
#                 Path('/projects/wakedynamics/orybchuk/ldm-3d/logs/2022-12-26T07-28-14_rank_geo_raaw_kl1_0020/images/test/postprocessed/'),
#                 Path('/projects/wakedynamics/orybchuk/ldm-3d/logs/2022-12-26T07-36-10_rank_geo_raaw_kl1_0030/images/test/postprocessed/'),
#                 Path('/projects/wakedynamics/orybchuk/ldm-3d/logs/2022-12-26T07-55-55_rank_geo_raaw_kl1_0040/images/test/postprocessed/'),
#                 Path('/projects/wakedynamics/orybchuk/ldm-3d/logs/2022-12-26T07-55-55_rank_geo_raaw_kl1_0050/images/test/postprocessed/'),
#                 Path('/projects/wakedynamics/orybchuk/ldm-3d/logs/2022-12-26T08-08-57_rank_geo_raaw_kl1_0060/images/test/postprocessed/'),
#                 Path('/projects/wakedynamics/orybchuk/ldm-3d/logs/2022-12-26T08-08-57_rank_geo_raaw_kl1_0070/images/test/postprocessed/'),
#                 Path('/projects/wakedynamics/orybchuk/ldm-3d/logs/2022-12-26T08-26-49_rank_geo_raaw_kl1_0080/images/test/postprocessed/'),
#                 Path('/projects/wakedynamics/orybchuk/ldm-3d/logs/2022-12-26T08-26-49_rank_geo_raaw_kl1_0090/images/test/postprocessed/'),
#                 Path('/projects/wakedynamics/orybchuk/ldm-3d/logs/2022-12-26T08-26-49_rank_geo_raaw_kl1_0100/images/test/postprocessed/'),
#                 Path('/projects/wakedynamics/orybchuk/ldm-3d/logs/2022-12-26T08-26-49_rank_geo_raaw_kl1_0110/images/test/postprocessed/'),]
# rank_dirs = [Path('/projects/wakedynamics/orybchuk/ldm-3d/logs/2023-10-12T17-21-09_split-rank-geo-raaw-kl1_0000/images/test/postprocessed/'),
#              Path('/projects/wakedynamics/orybchuk/ldm-3d/logs/2023-10-12T17-37-20_split-rank-geo-raaw-kl1_0010/images/test/postprocessed/'),
#              Path('/projects/wakedynamics/orybchuk/ldm-3d/logs/2023-10-12T17-58-33_split-rank-geo-raaw-kl1_0020/images/test/postprocessed/'),
#              Path('/projects/wakedynamics/orybchuk/ldm-3d/logs/2023-10-12T18-14-42_split-rank-geo-raaw-kl1_0030/images/test/postprocessed/'),
#              Path('/projects/wakedynamics/orybchuk/ldm-3d/logs/2023-10-12T18-25-51_split-rank-geo-raaw-kl1_0040/images/test/postprocessed/'),
#              Path('/projects/wakedynamics/orybchuk/ldm-3d/logs/2023-10-12T18-25-52_split-rank-geo-raaw-kl1_0050/images/test/postprocessed/'),]
rank_dirs = list(Path('/projects/wakedynamics/orybchuk/ldm-3d/logs/').glob('*split-rank-geo-raaw*/images/test/postprocessed/'))
rank_dirs.sort()
```

```python
def unnorm(xp, xmin, xmax):
    '''
    Take data from [-1,1] back to the original values
    '''
    return (xp + 1)*0.5*(xmax-xmin)+xmin

umin, umax, vmin, vmax, wmin, wmax = 2.86975098, 12.5567627, -0.9810791, 4.91235352, -1.98095703, 2.5579834
```

```python
### Calculate rank once
## Helpful parameters
n_ens = 10
i0_assess = int((1800-360)/15)  # At met mast
i1_assess = int((1800-1005)/15) # Edge of scanning lidar
i2_assess = int(480/15)         # Beyond scanning lidar influence
i3_assess = int(1800/15)        # Turbine
j_assess = 64
n_obs_per_dir = 10

ranks = []

rmse_list_u0 = []
rmse_list_u1 = []
rmse_list_u2 = []
rmse_list_u3 = []

rmse_list_v0 = []
rmse_list_v1 = []
rmse_list_v2 = []
rmse_list_v3 = []

rmse_list_w0 = []
rmse_list_w1 = []
rmse_list_w2 = []
rmse_list_w3 = []


mae_list_u0 = []
mae_list_u1 = []
mae_list_u2 = []
mae_list_u3 = []

mae_list_v0 = []
mae_list_v1 = []
mae_list_v2 = []
mae_list_v3 = []

mae_list_w0 = []
mae_list_w1 = []
mae_list_w2 = []
mae_list_w3 = []

err_list_u0 = []
err_list_u1 = []
err_list_u2 = []
err_list_u3 = []

err_list_v0 = []
err_list_v1 = []
err_list_v2 = []
err_list_v3 = []

err_list_w0 = []
err_list_w1 = []
err_list_w2 = []
err_list_w3 = []

## Iterate through and open data
for rank_dir in rank_dirs:
    for obs_num in range(n_obs_per_dir):  # Iterate through observations
        # Open and arrange files
        gt_raw = np.load(Path(rank_dir, f'input_{str(obs_num).zfill(4)}.npy'))
        obs_raw = np.load(Path(rank_dir, f'conditioning_{str(obs_num).zfill(4)}.npy'))

        ens_raw = np.zeros((n_ens, gt_raw.shape[0], gt_raw.shape[1], gt_raw.shape[2], gt_raw.shape[3]))
        for i in range(n_ens):
            one_ens = np.load(Path(rank_dir, f'ens_{str(obs_num).zfill(4)}_{str(i).zfill(4)}.npy'))
            ens_raw[i,:,:,:,:] = one_ens.copy()
            
        # Un-normalize data
        gt = np.zeros_like(gt_raw)
        gt[0,:,:,:] = unnorm(gt_raw[0,:,:,:], umin, umax)
        gt[1,:,:,:] = unnorm(gt_raw[1,:,:,:], vmin, vmax)
        gt[2,:,:,:] = unnorm(gt_raw[2,:,:,:], wmin, wmax)
        
        ens = np.zeros_like(ens_raw)
        ens[:,0,:,:,:] = unnorm(ens_raw[:,0,:,:,:], umin, umax)
        ens[:,1,:,:,:] = unnorm(ens_raw[:,1,:,:,:], vmin, vmax)
        ens[:,2,:,:,:] = unnorm(ens_raw[:,2,:,:,:], wmin, wmax)
        
        # Calculate error
        err_u0 = ens[:,0,i0_assess,j_assess,:] - gt[0,i0_assess,j_assess,:]
        err_u1 = ens[:,0,i1_assess,j_assess,:] - gt[0,i1_assess,j_assess,:]
        err_u2 = ens[:,0,i2_assess,j_assess,:] - gt[0,i2_assess,j_assess,:]
        err_u3 = ens[:,0,i3_assess,j_assess,:] - gt[0,i3_assess,j_assess,:]
        
        err_v0 = ens[:,1,i0_assess,j_assess,:] - gt[1,i0_assess,j_assess,:]
        err_v1 = ens[:,1,i1_assess,j_assess,:] - gt[1,i1_assess,j_assess,:]
        err_v2 = ens[:,1,i2_assess,j_assess,:] - gt[1,i2_assess,j_assess,:]
        err_v3 = ens[:,1,i3_assess,j_assess,:] - gt[1,i3_assess,j_assess,:]
        
        err_w0 = ens[:,2,i0_assess,j_assess,:] - gt[2,i0_assess,j_assess,:]
        err_w1 = ens[:,2,i1_assess,j_assess,:] - gt[2,i1_assess,j_assess,:]
        err_w2 = ens[:,2,i2_assess,j_assess,:] - gt[2,i2_assess,j_assess,:]
        err_w3 = ens[:,2,i3_assess,j_assess,:] - gt[2,i3_assess,j_assess,:]
        
        err_list_u0.append(err_u0)
        err_list_u1.append(err_u1)
        err_list_u2.append(err_u2)
        err_list_u3.append(err_u3)
        
        err_list_v0.append(err_v0)
        err_list_v1.append(err_v1)
        err_list_v2.append(err_v2)
        err_list_v3.append(err_v3)
        
        err_list_w0.append(err_w0)
        err_list_w1.append(err_w1)
        err_list_w2.append(err_w2)
        err_list_w3.append(err_w3)
        
## Calculate statistics
# Reformat error into arrays
err_arr_u0 = np.array(err_list_u0)
err_arr_u1 = np.array(err_list_u1)
err_arr_u2 = np.array(err_list_u2)
err_arr_u3 = np.array(err_list_u3)

err_arr_v0 = np.array(err_list_v0)
err_arr_v1 = np.array(err_list_v1)
err_arr_v2 = np.array(err_list_v2)
err_arr_v3 = np.array(err_list_v3)

err_arr_w0 = np.array(err_list_w0)
err_arr_w1 = np.array(err_list_w1)
err_arr_w2 = np.array(err_list_w2)
err_arr_w3 = np.array(err_list_w3)

# Calculate MAE
mae_arr_u0 = np.mean(err_arr_u0, axis=(0,1))
mae_arr_u1 = np.mean(err_arr_u1, axis=(0,1))
mae_arr_u2 = np.mean(err_arr_u2, axis=(0,1))
mae_arr_u3 = np.mean(err_arr_u3, axis=(0,1))

mae_arr_v0 = np.mean(err_arr_v0, axis=(0,1))
mae_arr_v1 = np.mean(err_arr_v1, axis=(0,1))
mae_arr_v2 = np.mean(err_arr_v2, axis=(0,1))
mae_arr_v3 = np.mean(err_arr_v3, axis=(0,1))

mae_arr_w0 = np.mean(err_arr_w0, axis=(0,1))
mae_arr_w1 = np.mean(err_arr_w1, axis=(0,1))
mae_arr_w2 = np.mean(err_arr_w2, axis=(0,1))
mae_arr_w3 = np.mean(err_arr_w3, axis=(0,1))

# Calculate RMSE
n_samples = (np.array(err_list_u0).shape[0] * np.array(err_list_u0).shape[1])

rmse_arr_u0 = np.sum(np.square(err_arr_u0), axis=(0,1)) / n_samples
rmse_arr_u1 = np.sum(np.square(err_arr_u1), axis=(0,1)) / n_samples
rmse_arr_u2 = np.sum(np.square(err_arr_u2), axis=(0,1)) / n_samples
rmse_arr_u3 = np.sum(np.square(err_arr_u3), axis=(0,1)) / n_samples

rmse_arr_v0 = np.sum(np.square(err_arr_v0), axis=(0,1)) / n_samples
rmse_arr_v1 = np.sum(np.square(err_arr_v1), axis=(0,1)) / n_samples
rmse_arr_v2 = np.sum(np.square(err_arr_v2), axis=(0,1)) / n_samples
rmse_arr_v3 = np.sum(np.square(err_arr_v3), axis=(0,1)) / n_samples

rmse_arr_w0 = np.sum(np.square(err_arr_w0), axis=(0,1)) / n_samples
rmse_arr_w1 = np.sum(np.square(err_arr_w1), axis=(0,1)) / n_samples
rmse_arr_w2 = np.sum(np.square(err_arr_w2), axis=(0,1)) / n_samples
rmse_arr_w3 = np.sum(np.square(err_arr_w3), axis=(0,1)) / n_samples
```

```python
fig, ax = plt.subplots(1, 3, figsize=(6, 3), dpi=125, sharey=True)

heights = 15 * np.arange(64)

# ax[0].plot(rmse_arr_u3, heights, color='k', label='x=1800 m')
ax[0].plot(rmse_arr_u0, heights, color='C0', label='x=1440 m')
ax[0].plot(rmse_arr_u1, heights, color='C1', label='x=795 m')
ax[0].plot(rmse_arr_u2, heights, color='C2', label='x=480 m')

# ax[1].plot(rmse_arr_v3, heights, color='k')
ax[1].plot(rmse_arr_v0, heights, color='C0')
ax[1].plot(rmse_arr_v1, heights, color='C1')
ax[1].plot(rmse_arr_v2, heights, color='C2')

# ax[2].plot(rmse_arr_w3, heights, color='k')
ax[2].plot(rmse_arr_w0, heights, color='C0')
ax[2].plot(rmse_arr_w1, heights, color='C1')
ax[2].plot(rmse_arr_w2, heights, color='C2')


ax[0].legend(fontsize=8)
ax[0].set_ylim(-10, 480)

for axs in ax:
    axs.axhline(56.5, color='k', alpha=0.2)
    axs.axhline(183.5, color='k', alpha=0.2)

ax[0].text(0.95,
      0.22,
      'Rotor\nDisk',
      fontsize=10,
      transform = ax[0].transAxes,
      horizontalalignment='right',
      color='black',
      alpha=0.30)

ax[0].set_ylabel('Height [m]')
ax[0].set_xlabel('$u$ RMSE [m s$^{-1}$]')
ax[1].set_xlabel('$v$ RMSE [m s$^{-1}$]')
ax[2].set_xlabel('$w$ RMSE [m s$^{-1}$]')

ax[0].set_xlim(0.0, 2.0)
ax[1].set_xlim(0.0, 0.6)
ax[2].set_xlim(0.0, 0.6)

for axs, letter in zip(ax.flatten(), list(string.ascii_lowercase)):
    axs.text(0.99,
          0.025,
          f'({letter})',
          fontsize=8,
          transform = axs.transAxes,
          horizontalalignment='right',
          color='black')

# plt.savefig('../figs/rmse_profiles.pdf', bbox_inches='tight')

plt.show()
```
